# Parabolic Stop and Reverse 

```python
import matplotlib as mpl

from mintalib.indicators import SAR, HMA
from mintalib.samples import sample_prices

mpl.rc("figure", figsize=(12, 9))
```

```python
prices = sample_prices().assign(hma=HMA(20), sar=SAR())
prices.info()
```

```python
prices.tail(250).reset_index().plot(y=["close", "hma", "sar"], style=["k", "b", "g."]);
```
