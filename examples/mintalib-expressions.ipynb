# Polars Expression Factory Functions


```python
import polars as pl
import mintalib.expressions as ta
```

```python
from mintalib.samples import sample_prices

prices = sample_prices(backend="polars")
prices
```

```python
prices.select(
    ta.ema(20).alias("ema20")
)
```

```python
prices.select(
    ta.atr(14).alias("atr")
)

```

```python
prices.select(
    ta.macd()
)
```

```python
prices.select(
    ta.macd().struct.unnest()
)
```

```python
# Expression can be piped into other expression factory functions.
# The first expression argument is passed as `src` to the next function.

prices.select(
    ta.ema(20).pipe(ta.roc, 1)
)
```

```python
prices.select(
    ta.macd().struct.unnest(),
    sma=ta.sma(50),
    atr=ta.atr(14),
    trend=ta.ema(50).pipe(ta.roc, 1)
)
```
