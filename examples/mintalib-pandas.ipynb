```python
import numpy as np
import pandas as pd

import yfinance as yf
```

```python
prices = yf.Ticker('AAPL').history('5y')
prices = prices.rename(columns=str.lower).rename_axis(index=str.lower)
prices

```

```python
np.asarray(prices['close'], dtype=None)
```

```python
from mintalib.indicators import EMA, SMA, ROC, RSI, EVAL, LOG

# compute and append indicators to prices
result = prices.assign(
    sma50 = SMA(50),
    sma200 = SMA(200),
    rsi = RSI(14),
    slope = ROC(1) @ EMA(20) @ LOG(),
    uptrend = EVAL("sma50 > sma200")
)

result
```
