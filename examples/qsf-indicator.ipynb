# Quadratic Series Forecast (Quadratic Regression)

```python
import matplotlib.pyplot as plt

from mintalib.indicators import HMA, TSF, QSF, DEMA, TEMA
from mintalib.samples import sample_prices

plt.rc("figure", figsize=(12, 9))
```

```python
PERIOD = 50

prices = sample_prices().assign(
    tsf=TSF(PERIOD),
    qsf=QSF(PERIOD),
    hma=HMA(PERIOD),
    dema=DEMA(PERIOD),
    tema=TEMA(PERIOD)
)

prices.info()
```

```python
prices.tail(250).plot(
    y=["close", "qsf", "tsf"],
    title="QSF vs TSF",
    use_index=False,
);
```

```python
prices.tail(250).plot(
    y=["close", "qsf", "hma"],
    title="QSF vs HMA",
    use_index=False
);
```

```python
prices.tail(250).plot(
    y=["close", "qsf", "dema"],
    title="QSF vs DEMA",
    use_index=False,
);
```

```python
prices.tail(250).plot(
    y=["close", "qsf", "tema"],
    title="QSF vs TEMA",
    use_index=False,
);
```
