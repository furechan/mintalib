# Time Series Forecast (Linear Regression)

```python
import matplotlib.pyplot as plt

from mintalib.indicators import HMA, TSF, DEMA, TEMA
from mintalib.samples import sample_prices

plt.rc("figure", figsize=(12, 9))
```

```python
PERIOD = 50

prices = sample_prices().assign(
    tsf=TSF(PERIOD),
    hma=HMA(PERIOD),
    dema=DEMA(PERIOD),
    tema=TEMA(PERIOD)
)

prices.info()
```

```python
prices.tail(250).plot(
    y=["close", "tsf", "hma"],
    title="TSF vs HMA",
    use_index=False
);
```

```python
prices.tail(250).plot(
    y=["close", "tsf", "dema"],
    title="TSF vs DEMA",
    use_index=False
);
```

```python
prices.tail(250).plot(
    y=["close", "tsf", "tema"],
    title="TSF vs TEMA",
    use_index=False
);
```
