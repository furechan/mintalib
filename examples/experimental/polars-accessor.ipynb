# Polars Accessor

> **Experimental** — consider using `mintalib.expressions` for polars-native workflows.

This notebook demonstrates the `mintalib.polars` module — a `ts` accessor registered on polars series, dataframes, and expressions.

Import `mintalib.polars` to activate the accessor. Two convenience expressions are provided: `CLOSE` (`pl.col("close")`) and `OHLC` (`pl.struct(["open", "high", "low", "close"])`). Indicators expecting a prices input should be called on a struct. Multi-column outputs like `macd` return a polars struct; use `.struct.unnest()` to expand into separate columns.

```python
import polars as pl

from mintalib.polars import CLOSE, OHLC # noqa 
```

```python
from mintalib.samples import sample_prices

prices = sample_prices(backend="polars")
prices
```

```python
prices.ts.ema(20)
```

```python
prices.ts.atr()
```

```python
prices.ts.macd()
```

```python
prices.select(CLOSE.ts.ema(20))
```

```python
prices.select(OHLC.ts.atr())
```

```python
prices.select(
    pl.col('close').ts.macd().struct.unnest()
)
```
