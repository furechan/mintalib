# Chart Example (based on mplfinance) 

```python
# requires mplfinance

from __helper__ import Chart

from mintalib.indicators import SMA, RSI
from mintalib.samples import sample_prices
```

```python
prices = sample_prices()
prices
```

```python
chart = Chart(title="Example", figsize=(16, 9)).add(
    SMA(20)
) .add(
    SMA(50)
).add(
    RSI(), panel="next"
)

chart.plot(prices, max_bars=250)
```
