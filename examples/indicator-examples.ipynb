# Using mintalib indicators

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

from mintalib.samples import sample_prices
from mintalib.indicators import ROC, SMA, EMA, RSI, MACD, TRANGE, ATR, SLOPE, EVAL

mpl.rc("figure", figsize=(12, 9))
```

```python
prices = sample_prices()
prices.info()
```

```python
prices = prices.assign(
    sma20 = SMA(20),
    sma50 = SMA(50)
)

prices.tail(250).plot(
    y=["close", "sma20", "sma50"],
);
```

```python
prices = prices.assign(
    slope = SLOPE(20),
)

axes = prices.tail(250).plot(
    y=["close", "slope"],
    subplots=[['close']],
)

ax = axes[-1]

gs = ax.get_gridspec()
gs.set_height_ratios([3, 1])
gs.update(hspace=0.0)
```

```python
prices = prices.assign(
    atr = ATR()
)

axes = prices.tail(250).plot(
    y=["close", "atr"],
    subplots=[['close']],
)

ax = axes[-1]

gs = ax.get_gridspec()
gs.set_height_ratios([3, 1])
gs.update(hspace=0.0)
```

```python
prices.assign(
    rsi=RSI(),
    sma20=SMA(20),
    sma50=SMA(50),
    sma200=SMA(200),
    slope20=SLOPE(20),
    pos=EVAL("sma50 > sma200 and slope20 > 0.0"),
).tail()
```

```python
macd = MACD() @ prices
macd
```

```python
fig, ax = plt.subplots()

# macd.tail(250).plot(y = "macdhist", style="b", kind="bar", xticks=[], ax=ax)
macd.tail(250).plot(y = ["macd", "macdsignal"], style=["k", "b"], kind="line", ax=ax)

```
