# Using indicators with pandas


```python
import numpy as np
import pandas as pd

from mintalib.samples import sample_prices
from mintalib.indicators import EMA, SMA, ROC, RSI, EVAL, LOG
```

```python
prices = sample_prices()

result = prices.assign(
    sma50 = SMA(50),
    sma200 = SMA(200),
    rsi = RSI(14),
    slope = ROC(1) @ EMA(20) @ LOG(),
    uptrend = EVAL("sma50 > sma200")
).loc[:, "sma50":]

result
```
