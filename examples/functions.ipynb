# Using Functions

This notebook demonstrates the `mintalib.functions` module — the simplest interface, providing plain calculation functions.

Functions are named in **lower case** (e.g. `sma`, `ema`, `macd`). The first parameter is either `prices` (a pandas/polars DataFrame with columns `open`, `high`, `low`, `close`, `volume`) or `series` (a pandas/polars series or numpy array).

```python
import mintalib.functions as ta
```

```python
from mintalib.samples import sample_prices

prices = sample_prices()
prices
```

```python
ta.sma(prices, period=20)
```

```python
ta.macd(prices, 12, 26, 9)
```

```python
ta.atr(prices, 14)

```
