# Hull Moving Average

```python
import matplotlib.pyplot as plt

from mintalib.indicators import HMA, DEMA, TEMA
from mintalib.samples import sample_prices

plt.rc("figure", figsize=(12, 9))
```

```python
PERIOD = 50

prices = sample_prices().assign(
    hma=HMA(PERIOD),
    dema=DEMA(PERIOD),
    tema=TEMA(PERIOD)
)

prices.info()
```

```python
prices.tail(250).plot(
    y=["close", "hma", "dema"],
    title="HMA vs DEMA",
    use_index=False
);
```

```python
prices.tail(250).plot(
    y=["close", "hma", "tema"],
    title="HMA vs TEMA",
    use_index=False
);
```
