# Keltner Channel

```python
import matplotlib as mpl

from mintalib.functions import KELTNER
from mintalib.samples import sample_prices

mpl.rc("figure", figsize=(12, 9))
```

```python
prices = sample_prices()
prices = prices.join(KELTNER(prices))
prices.info()
```

```python
prices.tail(250).plot(
    y=["close", "middleband", "upperband", "lowerband"],
    style=["k", "b--", "b:", "b:"],
);
```
